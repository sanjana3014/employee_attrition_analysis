As we need to find the correlation between all the numerical features and later visualize it, we chose a **'Heatmap'**.

```python
import pandas as pd

employee_attrition_data = pd.read_csv('dvi-a2-ps8-data.csv')

print(employee_attrition_data)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = employee_attrition_data.select_dtypes(include=[np.number]).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
            cbar_kws={"shrink": .75}, fmt=".2f", square=True, linewidths=.5)
plt.title('Correlation Matrix for Numerical Features in Employee Attrition Dataset')
plt.tight_layout()
plt.show()
```

**1. Summarize your findings from the visual:**

The heatmap indicates several key points about the relationships between numerical features:

Age seems to have a strong positive correlation with TotalWorkingYears, which is intuitive as older employees are likely to have worked longer.
MonthlyIncome is highly correlated with TotalWorkingYears as well, suggesting that income increases with work experience.
PercentSalaryHike does not show a strong correlation with other features, indicating that salary increases may not be strongly influenced by the other numerical factors measured here.
NumCompaniesWorked does not show a strong positive or negative correlation with TotalWorkingYears, which may imply that the diversity of company experience does not necessarily increase with the number of years worked.
Lower correlations across the board with DistanceFromHome suggest that it may not be strongly related to the other measured factors.


**2. The reason for selecting the chart type you did**

A heatmap is selected for its ability to visually represent the strength and direction of correlations between multiple numerical features simultaneously. The color intensity aids in quickly identifying which pairs of features are most strongly related.

**3. Mention the pre-attentive attributes used (at least 2):**

1. **Color Saturation:** The intensity of the red color on the heatmap visually represents the strength of a positive correlation, while the intensity of the blue color represents the strength of a negative correlation. This quickly draws the eye to stronger relationships.
2. **Numerical Annotations:** The actual correlation values are annotated within the heatmap cells. These annotations provide an immediate and precise understanding of the correlation strength without needing to interpret the color saturation alone..

**4. Mention the gestalt principles used (at least 2)**
Similarity: The heatmap uses colors similarly to group features with positive correlations separately from those with negative correlations.
Proximity: Correlated features are positioned next to each other, making it easier to perceive their relationship due to their spatial closeness.

```python
import matplotlib.pyplot as plt
import seaborn as sns

employee_attrition_data['Attrition'] = employee_attrition_data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

department_attrition = employee_attrition_data.groupby('Department')['Attrition'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Attrition', data=department_attrition, palette='coolwarm')
plt.title('Department-wise Attrition Rate')
plt.xlabel('Department')
plt.ylabel('Attrition Rate')
plt.show()
```

**1. Summarise your findings from the visual.**
The bar chart suggests that the Human Resources department has the highest attrition rate, followed by the Sales department, with Research & Development having the lowest among the three. This could indicate department-specific factors influencing employee turnover, such as job satisfaction, work environment, or growth opportunities.

**2. The reason for selecting the chart type**
A bar chart is ideal for comparing numerical data across different categories. It clearly shows differences in attrition rates across departments, allowing for immediate visual comparisons.

**3. Mention the pre-attentive attributes used (at least 2)**

**Color:** Each bar is colored differently to distinguish between departments easily.

**Length of Bars:** The varying lengths of the bars represent different attrition rates, which quickly indicate the magnitude of attrition in each department.

**4. Mention the gestalt principles used (at least 2)**

**Proximity:** Bars are positioned next to each other, which helps compare the departments directly based on their proximity.

**Similarity:** Bars are similar in shape and orientation, which allows the viewer to understand they represent the same type of data across different categories.

<h4>Identify the top 5 job roles with highest attrition rate.<br>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data_path = 'dvi-a2-ps8-data.csv'
employee_attrition_data = pd.read_csv(data_path)

employee_attrition_data['Attrition'] = employee_attrition_data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

attrition_rate_by_role = employee_attrition_data.groupby('JobRole')['Attrition'].mean().sort_values(ascending=False)

top_roles = attrition_rate_by_role.head(5)

plt.figure(figsize=(10, 6))
top_roles.plot(kind='bar', color='skyblue')
plt.title('Top 5 Job Roles with Highest Attrition Rate')
plt.xlabel('Job Role')
plt.ylabel('Attrition Rate')
plt.show()
```

**1. Summarise your findings from the visual.**

The bar chart illustrates that Sales Representatives have the highest attrition rate, followed by Laboratory Technicians, Human Resources, Sales Executives, and Research Scientists. This suggests that roles with direct sales responsibilities or those involving frequent interaction with products and customers might be experiencing higher stress or job dissatisfaction, leading to higher turnover.

**2. The reason for selecting the chart type you did**

A **bar chart** is ideal for comparing the attrition rate across different job roles, as it allows for an immediate visual comparison of the height of the bars. It makes it clear which roles have higher attrition rates, facilitating quick and effective analysis.

**3. Mention the pre-attentive attributes used (at least 2)**

**Color:** The consistent color across bars focuses attention on the height differences rather than differences in color, allowing for an unbiased comparison of the data.

**Length:** The length of the bars is a direct representation of the attrition rates, making it easy to compare values at a glance.

**4. Mention the gestalt principles used (at least 2)**

**Proximity:** Bars are placed next to one another, which indicates they are part of the same category (job roles) and should be compared with each other.

**Similarity:** The similar shape and color of bars imply that they are comparable entities representing different instances of the same measure (attrition rate).

<h4>Identify the top 2 highly correlated features.Analyse(use a single visual), how they relate to "number of years in the company" and their combined effect on attrition.<br>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_path = 'dvi-a2-ps8-data.csv'
employee_attrition_data = pd.read_csv(data_path)

employee_attrition_data['Attrition'] = employee_attrition_data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

numeric_df = employee_attrition_data.select_dtypes(include=[np.number])

correlation_matrix = numeric_df.corr()

top_correlated_features = correlation_matrix['YearsAtCompany'].drop('YearsAtCompany').abs().sort_values(ascending=False).head(2).index.tolist()

sns.pairplot(employee_attrition_data, vars=top_correlated_features + ['YearsAtCompany'], hue='Attrition', palette='coolwarm')
plt.suptitle('Pairplot of Highly Correlated Features with YearsAtCompany and Attrition', verticalalignment='top')
plt.tight_layout()
plt.show()
```

**1. Summarise your findings from the visual.**

The pairplot reveals potential relationships between the 'TotalWorkingYears', 'MonthlyIncome', and 'YearsAtCompany', with distinct patterns when considering attrition (denoted by different colors). There seems to be a positive correlation between 'TotalWorkingYears' and 'MonthlyIncome', as well as between 'TotalWorkingYears' and 'YearsAtCompany', indicating that employees with more working years tend to have a higher income and have been longer with the company. Furthermore, the plot suggests that attrition is more common among employees with fewer total working years and less time with the company, as seen by the concentration of attrition points in those areas.

**2. The reason for selecting the chart type you did**

A pairplot is chosen for its ability to simultaneously display the distribution of single variables and the relationships between two variables. It allows for an immediate visual comparison of how different features relate to each other and to the 'Attrition' status.

**3. Mention the pre-attentive attributes used (at least 2)**

**Color:** Different colors are used to distinguish between employees who have left (attrition = 1) and those who have stayed (attrition = 0).

**Shape:** Different marker shapes can help to quickly differentiate between two groups in scatter plots. Although the shapes are not visible in the provided image, they would typically be set in the markers parameter of pairplot

**4. Mention the gestalt principles used (at least 2)**

**Proximity:** Points that are close to each other within each scatter plot panel suggest a correlation or relationship between the variables.

**Similarity:** Points that share color (and possibly shape) are perceived as being part of the same group (either attrited or not).

 **How does the rate of attrition compare across different age groups, and what age group has the highest attrition rate?**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dvi-a2-ps8-data.csv')

plt.figure(figsize=(10, 5))
sns.histplot(df[df['Attrition'] == 'Yes']['Age'], bins=range(18, 61, 1), kde=False, color='red')
plt.title('Age Distribution of Attrition')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**1. Summarise your findings from the visual.**

The histogram indicates that attrition is most prevalent among employees in their late 20s to early 30s, with a significant peak around the age of 29 or 30. There are smaller peaks in attrition for employees in their mid-30s and early 40s. The frequency of attrition appears to decrease for employees beyond their mid-40s.

**2. The reason for selecting the chart type you did**
    
The histogram was chosen for its effectiveness in showing the distribution of a single numerical variable (age) and highlighting the frequency at which each value (or range of values) occurs. It helps to quickly identify the age groups with the highest counts of attrition.

**3. Mention the pre-attentive attributes used (at least 2)**
                                                
Color: The red color of the bars stands out against the neutral background, drawing immediate attention to the data.
Height: The height of the bars indicates the count of attrition events, allowing for quick comparison between different age groups.

**4. Mention the gestalt principles used (at least 2)**

Proximity: Bars that are adjacent to each other suggest a relationship in the sequence of age groups and facilitate comparison between neighboring age ranges.
Similarity: All the bars are similar in color and shape, suggesting they are part of the same data set, which in this case represents the count of attrition events across different ages.
